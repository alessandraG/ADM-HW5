```python
import collections
import math
def Bfs(G,root,destination):
    step=0
    seen=set([root])
    queue=collections.deque([(root,step)])
    while queue:
        vertex=queue.popleft()
        if vertex==destination:
            return vertex[1]
        else:
            step=vertex[1]+1
            neighbors=list(G[vertex[0]].keys())
            if neighbors==[]:
                continue
            if destination in neighbors:
                return step
            for i in neighbors:
                if i not in seen:
                    queue.append((i,step))                    
                    seen.add(i)
    return(math.inf)
```
